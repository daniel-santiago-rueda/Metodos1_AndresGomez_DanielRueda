Parte (a):

Según las notas de la clase, la segunda derivada se define como $$f''(x)=\frac{f(x+h)-2f(x)+f(x-h)}{h^2}$$

Así que se puede pensar en representar esta operación como una convuloción con máscara $M=[1, -2, 1]$.

De esta forma se tendría que $$D^2f(x_n)=\frac{1}{h^2}\sum_{m=-1}^{1}M[m+1]f(x_{n-m}).$$

Ahora, se calculará la derivada segunda de $f(x)=\frac{1}{\sqrt{1+e^{-x^2}}}$ mediante el uso de este kernel.

```python
x = np.linspace(-10, 10, 400)
function = lambda x: 1/(np.sqrt(1+np.exp(-x**2)))
h = 0.05

mascara = [1.0, -2.0, 1.0]

#Se calcula la derivada central en cada punto como una sumatoria de 
#la máscara de convolución evaluada en los puntos adyacentes
def SecondDerivative(x, f, h):
    sum = 0
    for m in range(-1, 2):
        mascara_elem = mascara[m+1]
        sum+= f(x-m*h)*mascara_elem
    return sum/(h**2)

second_derivative = SecondDerivative(x, function, h)

plt.plot(x, second_derivative, color='violet')
plt.title("Derivada segunda con kernel de convolución")
plt.xlabel("x")
plt.ylabel("f '' (x)")
plt.show()
```
